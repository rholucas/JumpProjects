```python
#top 5 players short name and wages.
top5_players = fifa[['short_name','wage_eur']]
```

```python
top5_players.head()
```

```python
#top 5 players short name and wages that are getting highest salaries. 
top5_players=top5_players.sort_values(by=['wage_eur'],ascending=False)
```

```python
top5_players.head()
```

```python
#  bar plot for top 5 players and their wages
plt.figure(figsize=(7,5))
plt.bar(list(top5_players['short_name'])[0:5],list(top5_players['wage_eur'])[0:5],color="g")
plt.show()
```
